## 4) standardization the data

```python
y = train["SalePrice"]
X = train[cols]
ss = StandardScaler()
X = ss.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
lr.score(X_train, y_train)
```

```python
lr.score(X_test,y_test)
```

```python
lr =lr.fit(X,y)
```

```python
lr.score(X,y)
```

```python
scores = cross_val_score(lr, X,y, cv=10)
```

```python
print(scores)
print(np.mean(scores))
```

## Ridge 

```python
ridge_alphas = np.logspace(0, 5, 200)

optimal_ridge = RidgeCV(alphas=ridge_alphas, cv=10)
optimal_ridge.fit(X, y)

print(optimal_ridge.alpha_)
```

```python
ridge = Ridge(alpha=optimal_ridge.alpha_)

ridge_scores = cross_val_score(ridge, X, y, cv=10)

print(ridge_scores)
print(np.mean(ridge_scores))
```

```python
predictions = ridge.predict(X)
```

```python
Xtes = ss.transform(test[cols]) 
```

```python
rg = ridge.fit(X,y)
print(rg.predict(Xtes))
```

## lasso

```python
optimal_lasso = LassoCV(n_alphas=500, cv=10, verbose=0)
optimal_lasso.fit(X, y)

print(optimal_lasso.alpha_)
```

```python
lasso = Lasso(alpha=optimal_lasso.alpha_)

lasso_scores = cross_val_score(lasso, X, y, cv=10)

print(lasso_scores)
print(np.mean(lasso_scores))
```

## Elastic Net

```python
l1_ratios = np.linspace(0.01, 1.0, 25)

optimal_enet = ElasticNetCV(l1_ratio=l1_ratios, n_alphas=100, cv=10,
                            verbose=0)
optimal_enet.fit(X, y)

print(optimal_enet.alpha_)
print(optimal_enet.l1_ratio_)
```

```python
enet = ElasticNet(alpha=optimal_enet.alpha_, l1_ratio=optimal_enet.l1_ratio_)

enet_scores = cross_val_score(enet, X, y, cv=10)

print(enet_scores)
print(np.mean(enet_scores))
```
